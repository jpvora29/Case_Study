```python
country_ids_list = list(ftball_country[ftball_country['country name'].isin(['Portugal', 'Hungary'])].loc[:, "country id"])
```

```python
common_data = m_details[m_details['team id'].isin(country_ids_list)]
common_match_no = int(common_data[common_data.duplicated(['match no'])]['match no'])
```

```python
final_data = m_stats[m_stats['match no'] == common_match_no].reset_index()
```

```python
match_no = final_data['match no'][0]
goal_score = final_data['goal score'][0]
play_stage = final_data['play stage'][0]
play_date = pd.to_datetime(final_data['play date']).dt.date[0]
results = final_data['results'][0]
```

```python
print(
    "match no:", match_no, '\n'
    "goal score:", goal_score, '\n'
    "play stage:", play_stage, '\n'
    "play date:", play_date, '\n'
    "results:", results
)
```

```python
new_players = pd.merge(players, ftball_country, left_on='team id', right_on='country id').drop(['country id', 'country abbr'], 
                                                                                               axis=1)
```

```python
new_players
```

```python
player_count_dict = {}

def final_data(new_players, goals) -> pd.DataFrame():
    player_count_dict = {}
    for player_id in goals['player id']:
        player_count_dict[player_id] = player_count_dict.get(player_id, 0) + 1

    # print(player_count_dict)

    for index, row in new_players.iterrows():
        player_id = row['player id']
        if player_id in player_count_dict:
            new_players.loc[index, 'count'] = player_count_dict[player_id]
    
    new_players['count'] = new_players['count'].astype(int)
    
    
    return new_players.loc[:, ['player id', 'country name', 'count', 'posi to play']]
  
```

```python
final_df = final_data(new_players, goals)
```
